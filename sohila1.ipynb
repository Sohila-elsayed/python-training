```python

list_nums = []
while True :
   nums = input( "enter numbers separated by (,) or done : " )
   if nums.lower() == 'done' :
      break
   N =nums.split (",")
   for i in N :
    list_nums.append(int (i))
asc = sorted(list_nums)
des =  sorted(list_nums,reverse = True)
print ("Ascending : " , asc)
print ("Descending : " , des)
```

```python
def modify_list(lst) :
  lst.append(100)
  lst.insert(2,999)
  for i in lst :
    if i == 50 :
     lst.remove(50)
  lst.pop()
  print ("list : ", lst)
modify_list([1 , 6 , 9 , 50])
```

```python
students = []
for i in range(5) :
  name = input ("enter name : ")
  grade = float(input("enter grade : "))
  students.append([name , grade])
max_grade = students[0]
min_grade = students[0]
for i in  students :
    if i[1] > max_grade[1] :
      max_grade = i
    if i[1] < min_grade[1] :
      min_grade = i
print ("student with high grade : ", max_grade[0] , "grade : ", max_grade[1] )
print ("student with low  grade : ", min_grade[0] , "grade : ", min_grade[1] )
```

```python
def modify_fruits(fruits):
  fruits.append("kiwi")
  fruits.insert(2,"gape")
  fruits[0] = fruits[0].upper()
  fruits.pop(1)
  print ("fruits list : ", fruits)
modify_fruits(["banana","apple","mango","cherry"])
```

```python
temp = [49 , 14 , 12 , 9 , 36 , 31]
count_high_deg = 0
count_low_deg = 0
for i in temp :
  if i > 30 :
    count_high_deg += 1
  elif i < 15 :
    count_low_deg += 1
  else :
    print("  ")
print ("Temperatures above 30 " , count_high_deg)
print ("Temperatures below 15  " , count_low_deg)
```

```python
shopping_list = []
while True :
  print ("1.Add  2.remove  3.sort  4.Exit")
  choice = input("choice : ")
  if choice == "1":
    item1 = input("enter item : ")
    shopping_list.append(item1)
    print (shopping_list)
  elif choice == "2" :
    item2 = input(" enter item : ")
    if item2 in shopping_list :
      shopping_list.remove(item2)
      print(shopping_list)
    else :
      print ("not found")
  elif choice == "3" :
    print (sorted(shopping_list))
  elif choice == "4"  :
    break
```
